# Canvas Exploration For Automated Mandala Creation 

```python
import numpy as np

n = 5000 
theta = np.arange(n+1)*2*np.pi/n

def Sin_n(n, a_n=1):
    return a_n*np.sin(n*theta)

def Cos_n(n, a_n=1):
    return a_n*np.cos(n*theta)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

r1 = 10 + Sin_n(2*np.pi*0.2)

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r1)
ax.grid(False)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
